```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
folder_path = r'D:\Code\RL-for-CS\files\experiment_results'
result_path = r'D:\Code\RL-for-CS\files\experiment_pictures'
# 获取指定文件夹下的所有文件名
file_names = os.listdir(folder_path)
# 获取所有文件名的前缀
data_names = [name[:-4] for name in file_names]
# 读取各个文件
datas = {name : pd.read_csv(os.path.join(folder_path, name+'.csv')) for name in data_names}
datas["reward"]
```

```python
""" 绘制训练过程中的各个图像 """
data_list1 = ["accuracy", "precision", "recall", "test_loss", "train_loss"]
data_list2 = ["actor_loss", "critic_loss", "alpha_loss", "reward"]
for data_name in data_names:
    data = datas[data_name]
    max_i = data.shape[0]
    if data_name in data_list1:
        max_step = 400000
    elif data_name in data_list2:
        max_step = 4000
    max_i = data[data['Step'] <= max_step].shape[0]
    x = data['Step'][:max_i]
    y = data['Value'][:max_i]
    plt.figure()
    plt.plot(x, y)
    plt.xlabel('Step')
    plt.ylabel(data_name)
    plt.savefig(os.path.join(result_path, data_name + '.png'), dpi=300)
```
